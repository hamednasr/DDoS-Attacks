```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = pd.read_csv('KDDTrain.csv')
data
```

```python
# Remove rows contain null values (in the case there are null values)
# In our data, please remove the second row on csv file before importing the file

rows, columns = np.where(pd.isnull(data))
data_filtered = data.drop(labels=rows)

print("There are %d rows having null values" %(len(data) - len(data_filtered)))

data_filtered
```

```python
# Remove duplicate rows

data_unduplicated = data_filtered.drop_duplicates()
print("There are %d duplicated rows" %(len(data_filtered) - len(data_unduplicated)))
data_unduplicated
```

```python
# Define feature vectors and labels

X, y = data_unduplicated.iloc[:, :-1], data_unduplicated.iloc[:, -1]
print("Shape of X: %s" %str(X.shape), "- Type of X: %s" %str(type(X)))
print("Shape of y: %s" %str(y.shape), "- Type of y: %s" %str(type(y)))
y
```

```python
# Classes of data

print("The number of original classes: %d" %len(y.unique()))
np.array(y).shape
```

```python
# Transform from 23 classes to 5 main classes of Internet attacks
# normal --> normal
# back, neptune, pod, smurf, teardrop --> DOS
# ipsweep, nmap, portsweep, satan --> Probing
# ftp_write, guess_passwd, imap, multihop, phf, spy, warezclient, warezmaster --> R2L
# buffer_overflow, loadmodule, perl, rootkit --> U2R

y_array = np.array(y)
y_encoded = np.zeros((y.shape))
for i in range(len(y_array)):
    if y_array[i] in ['normal']:
        y_array[i] = 'normal'
        y_encoded[i] = 0
    elif y_array[i] in ['back', 'land', 'neptune', 'pod', 'smurf', 'teardrop']:
        y_array[i] = 'DOS'
        y_encoded[i] = 1
    elif y_array[i] in ['ipsweep', 'nmap', 'portsweep', 'satan']:
        y_array[i] = 'probing'
        y_encoded[i] = 2
    elif y_array[i] in ['ftp_write', 'guess_passwd', 'imap', 'multihop', 'phf', 'spy', 'warezclient', 'warezmaster']:
        y_array[i] = 'R2L'
        y_encoded[i] = 3
    elif y_array[i] in ['buffer_overflow', 'loadmodule', 'perl', 'rootkit']:
        y_array[i] = 'U2R'
        y_encoded[i] = 4
y_df = pd.Series(y_array)
y_df.unique()
```

```python
# Check the number of each class

#for name in y_df.unique():
for name in y_df.unique():
    print("The number of samples of class %s: %d" %(str(name), len(X[y_array==name])))
y_encoded[:10]
```

```python
# Sort samples according to labels (optional)

data_new = X
data_new['Label'] = y_encoded
data_sorted = data_new.sort_values(by='Label')
data_sorted
```

```python
# Split the training set and testing set
data_array = data_sorted.to_numpy()
X_new, y_new = data_array[:, :-1], data_array[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size=0.2, random_state=0)
```

```python
# Feature selection
feature_selection = SelectKBest(chi2, k=10).fit(X_train, y_train)
X_train_new = feature_selection.transform(X_train)
X_test_new = feature_selection.transform(X_test)
feature_selection.get_support()
```

```python
################################## CASE 2 ################################
##################### ---- WITH NORMALIZATION STEP -----##############
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
X_train_scaled = std_scaler.fit_transform(X_train_new)
X_test_scaled = std_scaler.fit_transform(X_test_new)
X_train_scaled.shape
```

```python
## Neural Network - one hidden layer - With Normalization
hidden_layer_sizes_=[(12,),(25,),(50,),(100,),(200,),(300,),(400,)]
activation_=['identity', 'logistic', 'tanh', 'relu']

param_grid = [{'hidden_layer_sizes': hidden_layer_sizes_, 'activation': activation_}]

grid_search = GridSearchCV(MLPClassifier(), 
                           param_grid=param_grid, 
                           cv=5,
                           scoring='f1_weighted',
                           n_jobs=-1,
                          verbose=2)

grid_search.fit(X_train_scaled, y_train)
mlpc1_clf = grid_search.best_estimator_
mlpc1_clf
```

```python
mlpc1=MLPClassifier(hidden_layer_sizes=(300,),activation='logistic',max_iter=1000)
mlpc1.fit(X_train_scaled,y_train)
```

```python
y_train_pred = mlpc1.predict(X_train_scaled)
y_test_pred = mlpc1.predict(X_test_scaled)
```

```python
# check Accuaracy on Traing and Testing Data
train_accuracy_mlpc1 = mlpc1.score(X_train_scaled,y_train)
print("Train Accuaracy: " , train_accuracy_mlpc1)

test_accuracy_mlpc1 = mlpc1.score(X_test_scaled,y_test)
print("Test Accuaracy: " , test_accuracy_mlpc1)
```

```python
print(confusion_matrix(y_test, y_test_pred))
print("==================================")
print(classification_report(y_test, y_test_pred))
```

```python
## Neural Network - two hidden layers - With Normalization
hidden_layer_sizes_=[(50,200),(100,300),(100,500),(200,400),(200,600),(300,600),(300,800)]
activation_=['logistic', 'tanh', 'relu']

param_grid = [{'hidden_layer_sizes': hidden_layer_sizes_, 'activation': activation_}]

grid_search = GridSearchCV(MLPClassifier(), 
                           param_grid=param_grid, 
                           cv=5,
                           scoring='f1_weighted',
                           n_jobs=-1,
                          verbose=3)

grid_search.fit(X_train_scaled, y_train)
mlpc2_clf = grid_search.best_estimator_
mlpc2_clf
```

```python
mlpc2=MLPClassifier(hidden_layer_sizes=(200,600),max_iter=1000)
mlpc2.fit(X_train_scaled,y_train)
```

```python
y_train_pred2 = mlpc2.predict(X_train_scaled)
y_test_pred2 = mlpc2.predict(X_test_scaled)
```

```python
# check Accuaracy on Traing and Testing Data
train_accuracy_mlpc2 = mlpc2.score(X_train_scaled,y_train)
print("Train Accuaracy: " , train_accuracy_mlpc2)

test_accuracy_mlpc2 = mlpc2.score(X_test_scaled,y_test)
print("Test Accuaracy: " , test_accuracy_mlpc2)
```

```python
print(confusion_matrix(y_test, y_test_pred2))
print("==================================")
print(classification_report(y_test, y_test_pred2))
```
